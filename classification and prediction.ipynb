```python
%matplotlib inline  
from matplotlib import pyplot as plt
import pandas as pd

# import data
data = pd.read_csv("two_class_data.csv", header=0)

#explore and familiar with datasets
print(data.head())
print(data.tail())

# classified data
x = data['x']
y = data['y']
c = data['class']

# scatter plot mapc parameter used for adding color
plt.scatter(x, y, c=c)
plt.show()
```

```python
#import sklearn model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron



#define feature values and target values
#based on feature--> predict target
feature = data[['x', 'y']].values
target = data['class'].values


# split the data into 70% as training sets, 30% as test sets
x_train, x_test, y_train, y_test = train_test_split(feature, target, test_size=0.3, random_state=50)
print(x_train)
print(y_train)

# buid up model
model = Perceptron()

# training model by feeding training datasets
model.fit(x_train, y_train)

# prediction
results = model.predict(x_test)

```

```python
# plot the training results by using default method
plt.scatter(x_train[:, 0], x_train[:, 1], alpha=0.3)

# plot the training results by using 'marker'
plt.scatter(x_test[:, 0], x_test[:, 1], marker=',', c=y_test)

# label/annotate the result on the testing datasets(C1,C2)
for i, txt in enumerate(results):
    plt.annotate(txt, (x_test[:, 0][i], x_test[:, 1][i]))

plt.show()
```

```python
#print the accuracy
print(model.score(x_test, y_test))
```

```python
#improve the accuracy by using SVM

from sklearn.svm import SVC


# construct the model
model = SVC()


model.fit(x_train, y_train)


results = model.predict(x_test)


plt.scatter(x_train[:, 0], x_train[:, 1], alpha=0.3)


plt.scatter(x_test[:, 0], x_test[:, 1], marker=',', c=y_test)


for i, txt in enumerate(results):
    plt.annotate(txt, (x_test[:, 0][i], x_test[:, 1][i]))

plt.show()

```

```python
#print the accuracy
print(model.score(x_test, y_test))
```

```python
#Dimension reduction
#Principal component analysis(PCA)

from matplotlib import pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.decomposition import PCA


# import data
data = pd.read_csv('zoo.csv', header=0)

# define feature value and target values
feature = data.iloc[:, 1:17].values
target = data['type'].values

# reduce the componets into 2
pca = PCA(n_components=2)
feature_pca = pca.fit_transform(feature)


# split the datasets into 70% traing sets and 30% testing sets respectively
x_train, x_test, y_train, y_test = train_test_split(feature_pca, target, test_size=0.3, random_state=50)

# buiding up model
model = SVC()

# training model
model.fit(x_train, y_train)

# make prediction
results = model.predict(x_test)

print(model.score(x_test, y_test))


plt.scatter(x_train[:, 0], x_train[:, 1], alpha=0.3)


plt.scatter(x_test[:, 0], x_test[:, 1], marker=',', c=y_test)


for i, txt in enumerate(results):
    plt.annotate(txt, (x_test[:, 0][i], x_test[:, 1][i]))

plt.show()

```

```python
#k-nearest neighborsKNN
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# import data
data = pd.read_csv('three_class_data.csv', header=0)
print(data.head())
print(data.tail())

# fefine feature values and target values
feature = data[['x', 'y']].values
target = data['class'].values

# split the data into 70% training sets and 30% testing sets
x_train, x_test, y_train, y_test = train_test_split(feature, target, test_size=0.3, random_state=50)

# buiding up KNN model
model = KNeighborsClassifier()

# training by applying KNN model
model.fit(x_train, y_train)

# make prediction
results = model.predict(x_test)

print(model.score(x_test, y_test))

#data visualization
cm0 = plt.cm.Oranges
cm1 = plt.cm.Greens
cm2 = plt.cm.Reds
cm_color = ListedColormap(['red', 'yellow'])

x_min, x_max = data['x'].min() - .5, data['x'].max() + .5
y_min, y_max = data['y'].min() - .5, data['y'].max() + .5

xx, yy = np.meshgrid(np.arange(x_min, x_max, .1),
                     np.arange(y_min, y_max, .1))

Z0 = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 0]
Z1 = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z2 = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 2]

Z0 = Z0.reshape(xx.shape)
Z1 = Z1.reshape(xx.shape)
Z2 = Z2.reshape(xx.shape)

plt.contourf(xx, yy, Z0, cmap=cm0, alpha=.9)
plt.contourf(xx, yy, Z1, cmap=cm1, alpha=.5)
plt.contourf(xx, yy, Z2, cmap=cm2, alpha=.4)


plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, cmap=cm_color)
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, cmap=cm_color, edgecolors='black')
plt.show()
```
